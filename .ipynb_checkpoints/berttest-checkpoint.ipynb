```python
import numpy as np
from nltk.tokenize import RegexpTokenizer
import string
import os
import pandas as pd
import contractions as cont_lib
from tqdm import tqdm
import re
```

```python
df = pd.read_csv('twitter-2013train-A.txt',sep="\t",names=['ID','label','text'])
df
```

```python
#Cleaning the tweets
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase

def clean(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r'http\S+', ' ', text)
    text = re.sub(r'pic.twitter\S+', ' ', text)
    text = decontracted(text)
    text = re.sub(r'\(([^)]+)\)', " ", text)
    text = re.sub('@[\w]+','',text)
    text = re.sub(r'[^a-zA-Z ]+', ' ', text)
    text = re.sub(r' \w{1,2}_', ' ', text)
    text = re.sub('\s+',' ', text)
    return text
```

```python
for i in tqdm(df.itertuples()):
    df.at[i[0], 'tweet'] = clean(i[3])
```

```python
df
```

```python
data = df[['label','tweet']].copy()
```

```python
data
```

```python
def tokenize (text):
    tokens = [word.strip(string.punctuation) for word in RegexpTokenizer(r'\b[a-zA-Z][a-zA-Z0-9]{1,50}\b').tokenize(text)]
    return tokens
```

```python
from sklearn.model_selection import train_test_split
```

```python
train = pd.DataFrame(columns=['label', 'tweet'])
validation = pd.DataFrame(columns=['label', 'tweet'])
for label in data.label.unique():
    emotion_data = data[data.label == label]
    train_data, validation_data = train_test_split(emotion_data, test_size=0.3)
    train = pd.concat([train, train_data])
    validation = pd.concat([validation, validation_data])
```

```python
train
```

```python
validation
```

```python
train['tweet']=train['tweet'].apply(lambda x:tokenize(x))
```

```python
train
```

```python
train_data = np.array(train['tweet'])#np.ndarray()
train_x=train_data.tolist()
train_data_y = np.array(train['label'])#np.ndarray()
train_y=train_data_y.tolist()
```

```python
validation['tweet']=validation['tweet'].apply(lambda x:tokenize(x))
test_data = np.array(validation['tweet'])#np.ndarray()
test_x=test_data.tolist()
test_data_y = np.array(validation['label'])#np.ndarray()
test_y=test_data_y.tolist()
```

```python
bert_embedding = BertEmbedding('bert-base-uncased', sequence_length=30)                                   
model = CNN_LSTM_Model(bert_embedding)
```

```python
model = BiLSTM_Model()
model.fit(train_x, train_y, test_x, test_y)

# Evaluate the model
model.evaluate(test_x, test_y)

# Model data will save to `saved_ner_model` folder
model.save('saved_classification_model')

# Load saved model
loaded_model = BiLSTM_Model.load_model('saved_classification_model')
loaded_model.predict(test_x[:10])

# To continue training, compile the newly loaded model first
loaded_model.compile_model()
model.fit(train_x, train_y, test_x, test_y)
```
